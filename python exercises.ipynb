```python
def password_checker(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any (char in '!@#$%^&*()_+' for char in password):
        return False
    return True
password_checker('Hunter@10')
```

```python
class Character:
    def __init__(self,name,health,attack):
        self.name=name
        self.health=health
        self.attack=attack
    
    def display_info(self):
        print('Name : ',self.name)
        print('Health : ',self.health)
        print("attack : " ,self.attack)
    def special_attack(self):
        pass
    def attack_enemy(self,other):
        if other.health>0 and other.health>=self.attack:
            other.health-=self.attack
            print(self.name," hit ",other.name,'and deals ',self.attack,' damage!!!\nCurrent Health: ',other.health)
    def is_alive(self):
        if self.health>0:
            return True
        elif self.health==0:
            return False
class Warrior(Character):
    def __init__(self,name,health,attack,armor):
        super().__init__(name,health,attack)
        self.armor=armor
    def display_info(self):
        print('Name : ',self.name)
        print('Health : ',self.health)
        print("attack : " ,self.attack)
        print("armor : " ,self.armor)
    def special_attack(self):
        print("Warrior swings a massive sword!")

class Mage(Character):
    def __init__(self,name,health,attack,mana):
        super().__init__(name,health,attack)
        self.mana=mana
    def display_info(self):
        print('Name : ',self.name)
        print('Health : ',self.health)
        print("attack : " ,self.attack)
        print("mana : " ,self.mana)
    def special_attack(self):
        print("Mage casts Fireball!")

warrior=Warrior('Aryan',100,12,13)
mage=Mage('Aryan evil',110,24,20)
warrior.attack_enemy(mage)
mage.display_info()
warrior.special_attack()
mage.special_attack()
mage.is_alive()
```

```python
class Restaurant:
    def __init__(self, name, location, rating):
        self.name = name
        self.location = location
        self.rating = rating
        self.menu = []
        self.orders = []

    def add_menu_item(self, item):
        self.menu.append(item)

    def remove_menu_item(self, item_name):
        for item in self.menu:
            if item.name == item_name:
                self.menu.remove(item)
                break

    def update_price(self, item_name, new_price):
        for item in self.menu:
            if item.name == item_name:
                item.price = new_price
                break

    def display_menu(self):
        print(f"\n----- {self.name} MENU -----")
        for item in self.menu:
            item.display_item()

    def receive_order(self, order):
        self.orders.append(order)

    def display_orders(self):
        print("\nRestaurant Orders")
        for order in self.orders:
            order.display_order()

class MenuItem:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def display_item(self):
        print(f"{self.name} ({self.category}) - ₹{self.price}")

    def update_price(self, new_price):
        self.price = new_price

class Cart:
    def __init__(self):
        self.items = []
        self.quantities = []
        self.total_bill = 0

    def add_item(self, item, quantity):
        self.items.append(item)
        self.quantities.append(quantity)

    def remove_item(self, item_name):
        for i in range(len(self.items)):
            if self.items[i].name == item_name:
                self.items.pop(i)
                self.quantities.pop(i)
                break

    def calculate_total(self):
        self.total_bill = 0
        for i in range(len(self.items)):
            self.total_bill += self.items[i].price * self.quantities[i]
        return self.total_bill

    def clear_cart(self):
        self.items.clear()
        self.quantities.clear()
        self.total_bill = 0

    def copy(self):
        new_cart = Cart()
        new_cart.items = self.items.copy()
        new_cart.quantities = self.quantities.copy()
        new_cart.total_bill = self.calculate_total()
        return new_cart

    def display_cart(self):
        print("\nCart")
        for i in range(len(self.items)):
            print(f"{self.items[i].name} x {self.quantities[i]}")
        print(f"Total = ₹{self.calculate_total()}")

class Customer:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number
        self.cart = Cart()
        self.order_history = []

    def view_menu(self, restaurant):
        restaurant.display_menu()

    def add_to_cart(self, item, quantity):
        self.cart.add_item(item, quantity)

    def remove_from_cart(self, item_name):
        self.cart.remove_item(item_name)

    def place_order(self, restaurant):
        order = Order(self, restaurant, self.cart, "Preparing")
        restaurant.receive_order(order)
        self.order_history.append(order)
        self.cart.clear_cart()
        return order

    def view_order_history(self):
        print("\nOrder History")
        for order in self.order_history:
            order.display_order()

    def display_profile(self):
        print(f"Customer: {self.name}")
        print(f"Phone: {self.phone_number}")

class Order:
    def __init__(self, customer, restaurant, ordered_items, status):
        self.customer = customer
        self.restaurant = restaurant
        self.ordered_items = ordered_items.copy()
        self.total_amount = self.ordered_items.calculate_total()
        self.status = status

    def generate_bill(self):
        print(f"Total Amount = ₹{self.total_amount}")

    def update_status(self, new_status):
        self.status = new_status

    def display_order(self):
        print("\n-----------------------")
        print(f"Customer : {self.customer.name}")
        print(f"Restaurant : {self.restaurant.name}")
        print(f"Status : {self.status}")

        for i in range(len(self.ordered_items.items)):
            item = self.ordered_items.items[i]
            qty = self.ordered_items.quantities[i]
            print(f"{item.name} x {qty}")

        print(f"Bill : ₹{self.total_amount}")
        print("-----------------------")

#simulation of the restaurant management system
# Create Restaurant
restaurant = Restaurant("Dominos", "Pune", 4.5)

# Create Menu Items
pizza = MenuItem("Farmhouse Pizza", 350, "Pizza")
burger = MenuItem("Veg Burger", 150, "Burger")
coke = MenuItem("Coke", 50, "Drink")

# Add Menu Items
restaurant.add_menu_item(pizza)
restaurant.add_menu_item(burger)
restaurant.add_menu_item(coke)

# Create Customer
customer = Customer("Aryan", "9876543210")

# View Menu
customer.view_menu(restaurant)

# Add Items to Cart
customer.add_to_cart(pizza, 2)
customer.add_to_cart(coke, 3)

# Display Cart
customer.cart.display_cart()

# Place Order
order = customer.place_order(restaurant)

# Bill
order.generate_bill()

# Customer Order History
customer.view_order_history()

# Restaurant Orders
restaurant.display_orders()

# Cart after placing order
customer.cart.display_cart()
```
