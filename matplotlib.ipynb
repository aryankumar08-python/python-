```python
import pandas as pd
sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [120000, 135000, 128000, 145000, 160000, 170000],
    "Profit": [20000, 25000, 23000, 28000, 35000, 40000],
    "Expenses": [100000, 110000, 105000, 117000, 125000, 130000]
}
df=pd.DataFrame(sales_data)
fig,axes = plt.subplots(2,2)
axes[0,0].plot(df['Month'],df['Revenue'],marker='o',markersize=7,markerfacecolor='red',alpha=1)#linestyle='-.',linewidth=2,color='Red')#alpha is used for transperancy
axes[0,0].set_title("Revenue vs Month")
axes[0,1].plot(df['Month'],df['Profit'])
axes[0,1].set_title('Profit vs Month')
axes[1,1].plot(df['Month'],df['Expenses'])
axes[1,1].set_title('Expenses vs Month')

plt.show()

```
