```python
import pandas as pd
sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [120000, 135000, 128000, 145000, 160000, 170000],
    "Profit": [20000, 25000, 23000, 28000, 35000, 40000],
    "Expenses": [100000, 110000, 105000, 117000, 125000, 130000]
}
df=pd.DataFrame(sales_data)
fig,axes = plt.subplots(2,2,figsize=(5,3))
axes[0,0].plot(df['Month'],df['Revenue'],marker='o',markersize=7,markerfacecolor='red',alpha=1)#linestyle='-.',linewidth=2,color='Red')#alpha is used for transperancy
axes[0,0].set_title("Revenue vs Month")
axes[0,1].plot(df['Month'],df['Profit'])
axes[0,1].set_title('Profit vs Month')
axes[1,1].plot(df['Month'],df['Expenses'])
axes[1,1].set_title('Expenses vs Month')

plt.show()

```

```python
market_share={
    "Brand": ["Apple", "Samsung", "Xiaomi", "Others"],
    "Share": [32, 28, 20, 20]
}
df=pd.DataFrame(market_share)
fig,ax=plt.subplots()
ax.pie(
    df['Share'],
    labels=df['Brand'],
    autopct="1.1%%",
    explode=[0.1,0,0,0],
    textprops={'color': 'white'} 
)
ax.set_title('Brand Market Share',color='white')
plt.show()
fig.savefig('abc.png',transparent=True,dpi=300) #dots per inch 
#fig.suptitle("Company Sales Dashboard") uses to give the title to the figure instead od speciifc plot 
```

```python
sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Revenue": [120000,135000,128000,145000,160000,170000],
    "Profit": [20000,25000,23000,28000,35000,40000],
    "Expenses": [100000,110000,105000,117000,125000,130000]
}
df=pd.DataFrame(sales_data)
fig,axes=plt.subplots(2,2,figsize=(10,10))
axes[0,0].plot(df['Month'],df['Revenue'],label='Revenue')
axes[0,0].set_title('Month vs Revenue')
axes[0,0].set_xlabel('Month')
axes[0,0].set_ylabel('Revenue')
axes[0,0].legend()

axes[0,1].bar(df['Month'],df['Profit'])
axes[0,1].set_title('Month vs Profit')
axes[0,1].set_xlabel('Month')
axes[0,1].set_ylabel('Profit')


axes[1,0].plot(df['Month'],df['Expenses'],label='Expense')
axes[1,0].set_title('Month vs Expense')
axes[1,0].set_xlabel('Month')
axes[1,0].set_ylabel('Expense')
axes[1,0].legend()

axes[1,1].axis('off')
fig.suptitle("Sales Data")
axes[0,0].grid(True)
axes[1,0].grid(True)
fig.tight_layout() #helps to uncongest the plots and helps it from overlapping
fig.savefig('dashboard.png')
plt.show()
```
