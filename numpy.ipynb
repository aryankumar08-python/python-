```python
class Student:
    def __init__(self,name):
        self.name=name
    def getName(self):
        return self.name
    def __str__(self): #for normal audience
        return self.name
    def __repr__(self): #for developers 
        return 'Student(BMW)'
aryan=Student('Aryan')
print(aryan) #<__main__.Student object at 0x0000015E9B5E1850>
print(aryan.getName())
print(aryan)
```
