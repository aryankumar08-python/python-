```python
class Student:
    def __init__(self,name):
        self.name=name
    def getName(self):
        return self.name
    def __str__(self): #for normal audience
        return self.name
    def __repr__(self): #for developers 
        return 'Student(BMW)'
aryan=Student('Aryan')
print(aryan) #<__main__.Student object at 0x0000015E9B5E1850>
print(aryan.getName())
print(aryan)
```

```python
def grade(marks):
    if marks>=50:
        return 'Pass'
    elif marks<50:
        return 'Fail'
df=pd.DataFrame({
    'Marks':[40,60,85]
})
df['Marks'].apply(grade)
```

```python
def grade(marks):
    if marks>=150:
        return 'Pass'
    elif marks<150:
        return 'Fail'
def total_marks(row):
    return row['Maths']+row['Science']
df=pd.DataFrame({
    'Maths':[80,70,90],
    'Science':[90,60,95]
})
df['Total']=df.apply(total_marks,axis=1) #axis 1 means that we are going row wise 
df['Total'].apply(grade)
```

```python
import pandas as pd 
def result(row):
    if row['Maths']+row['Science']>=130:
        return 'Pass'
    else:
        return 'Fail'
df=pd.DataFrame({
    'Maths':[50,70,90],
    'Science':[40,80,60]
})
df['Result']=df.apply(result,axis=1)
df
```
